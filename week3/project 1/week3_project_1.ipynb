3. Outliers:


```python
import numpy as np
import pandas as pd
```

```python
# Example DataFrame
data = {'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 200],
        'Salary': [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 500000],
        'Rating': [4.2, 4.5, 4.8, 3.9, 4.0, 4.2, 4.4, 4.6, 4.9, 2.5]}

df = pd.DataFrame(data)
```

```python
# Calculate the IQR for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Define a threshold to identify potential outliers
threshold = 1.5

# Identify outliers
outliers = ((df < (Q1 - threshold * IQR)) | (df > (Q3 + threshold * IQR))).any(axis=1)

# Display rows with potential outliers
print("Rows with potential outliers:")
print(df[outliers])
```

```python
# Remove rows with outliers
df_no_outliers = df[~outliers]
```

```python
# Replace outliers with the median value
df_adjusted = df.copy()
df_adjusted[outliers] = df.median()
```

```python
# Display the resulting DataFrame after handling outliers
print("\nDataFrame without outliers:")
print(df_no_outliers)
print("\nDataFrame with adjusted outliers:")
print(df_adjusted)
```

Function to check for special characters

```python

def contains_special_characters(text):
    return any(char.isalnum() or char.isspace() for char in text)
```

Each element in the 'Location' column

```python
 df['Contains_Special_Characters'] = np.vectorize(contains_special_characters)(df['Location'])
```

 Display rows with special characters

```python
rows_with_special_chars = df[df['Contains_Special_Characters']]
print("Rows with Special Characters:")
print(rows_with_special_chars[['Location', 'Contains_Special_Characters']])
```

Function to clean special characters

```python
def clean_special_characters(text):
     return ''.join(char for char in text if char.isalnum() or char.isspace())
```

Apply the function to clean the 'Location' column

```python
df['Cleaned_Location'] = np.vectorize(clean_special_characters)(df['Location'])
```

Display the cleaned DataFrame

```python
print("\nCleaned DataFrame:")
print(df[['Location', 'Cleaned_Location']])
```

13. Location Accuracy

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/Data-cleaning-for-beginners-using-pandas.csv")
```

Create a sample DataFrame

```python
data = {'Location': ['New York', 'NYC', 'Australia', 'Aus', 'India' ,'In']}
df = pd.DataFrame(data)
```

Create a reference list of valid locations

```python
valid_locations = ['New York', 'NYC', 'Australia', 'Aus', 'India' ,'In']
```

Assess accuracy and correct entries

```python
from difflib import get_close_matches

df['Corrected_Location'] = df['Location'].apply(lambda x: get_close_matches(x, valid_locations, n=1)[0] if x not in valid_locations else x)
```

Assess accuracy

```python
accuracy = (df['Location'] == df['Corrected_Location']).mean() * 100
print(f"Accuracy: {accuracy:.2f}%")
```

Display the DataFrame with corrected locations

```python
print(df)
```

```python
df['Standardized_Rating'] = (df['Rating'] - mean_rating) / std_rating
```

```python
print("Mean of Standardized Ratings:", mean_standardized)
print("Standard Deviation of Standardized Ratings:", std_standardized)
```
